```python
# Google pandas import
import pandas as pd

# Reading the datasets
red_wine_quality = pd.read_csv('./data/wine-quality/winequality-red.csv', sep=';')
white_wine_quality = pd.read_csv('./data/wine-quality/winequality-white.csv', sep=';')
wine_chemistry = pd.read_csv('./data/wine/wine.data', sep=',', header=None, names=[
    "Alcohol", "Malic acid", "Ash", "Alkalinity of ash", "Magnesium", 
    "Total phenols", "Flavonoids", "Nonflavonoid phenols", "Proanthocyanins", 
    "Colour intensity", "Hue", "OD280/OD315 of diluted wines", "Proline"
])

# Target Column Identification
# For the red wine dataset
red_target_col = 'quality'
red_target_type = "Numeric" if pd.api.types.is_numeric_dtype(red_wine_quality[red_target_col]) else "Categorical"

# For the white wine dataset
white_target_col = 'quality'
white_target_type = "Numeric" if pd.api.types.is_numeric_dtype(white_wine_quality[white_target_col]) else "Categorical"

# For the wine chemistry dataset
chemistry_target_col = 'Alcohol'
chemistry_target_type = "Numeric" if pd.api.types.is_numeric_dtype(wine_chemistry[chemistry_target_col]) else "Categorical"

# Displaying the results
print(f"Dataset: winequality-red.csv\nTarget Column: {red_target_col}\nData Type: {red_target_type}\n")
print(f"Dataset: winequality-white.csv\nTarget Column: {white_target_col}\nData Type: {white_target_type}\n")
print(f"Dataset: wine.data\nTarget Column: {chemistry_target_col}\nData Type: {chemistry_target_type}")
```

```python
# Part 2
#
# None of the data sets have missing values.
# White Wine Quality was chosen for the next
# steps.

df_num = white_wine_quality
x_train = df_num.drop('pH', axis=1)
y_train = df_num['pH']
```

```python
# Part 3

from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, n_jobs=-1)
rf.fit(x_train, y_train)
```

```python
r2 = rf.score(x_train, y_train)
print(f"{r2:.4f}")
```

### Comment on R2 value
The RF Score on the White Wine Quality data set (0.9577) indicates a very strong relationship in the DS's values, this means other quality metrics for wines with incomplete values are able to be predicted with high confidence.
